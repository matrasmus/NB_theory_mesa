```python
class SocialAgent(Agent):
    """Agent with personal network and state."""    
    def __init__(self, unique_id, model, n_nodes_int_belief, connected_agents, remaining_agents, beta_values, beta_soc, rep, extNet, network, conversion, weight_ext):

        super().__init__(unique_id, model)
        
        self.unique_id = unique_id
        #self.runs = runs
        self.n_nodes = n_nodes_int_belief
        self.connected_agents = connected_agents
        self.beta_intPers = beta_values['intPers']
        self.beta_intSoc = beta_values['intSoc']
        self.beta_intSoc_dynamic = beta_soc
        self.beta_ext = beta_values['ext']
        self.extNet = extNet
        self.weight_ext = weight_ext
        self.fullConnNet = 0 #here, a full network will be stored
        #predefined for equal networks at the start
        self.remaining_agents = remaining_agents
        self.pers_belief_network = self.create_belief_network(self.unique_id ,self.n_nodes, self.connected_agents, self.remaining_agents, weight_int_pers = 0.4, weight_int_soc = 1.0)
        self.rep = rep
        self.testnet = 'nan'
        self.conversion = conversion
        self.conversion_function = 0
        
        
        
        

    def create_belief_network(self, unique_id, n_nodes, connected_agents, remaining_agents, weight_int_pers, weight_int_soc):
        
        all_connections = connected_agents + remaining_agents

        
        seed = os.getpid() + random.randint(1, 100)
        rng = default_rng(seed)

        G = nx.DiGraph()

        for i in range(n_nodes):
            if i == 0: 
                G.add_node(f'agent{unique_id}_focal_belief', value=rng.choice(np.linspace(-1, 1, 7))) 
            else:
                G.add_node(f'agent{unique_id}_belief_{i}', value=rng.choice(np.linspace(-1, 1, 7)))

        for c in all_connections:
            G.add_node(f'agent{unique_id}_soc_belief_{c}', value=rng.choice(np.linspace(-1, 1, 7)))
            

        focal_node = f'agent{unique_id}_focal_belief'

        for i in range(1, n_nodes):
            G.add_edge(focal_node, f'agent{unique_id}_belief_{i}', weight=weight_int_pers)  
            G.add_edge(f'agent{unique_id}_belief_{i}', focal_node, weight=weight_int_pers)
            for j in range(1, n_nodes):
                if i != j:
                    G.add_edge(f'agent{unique_id}_belief_{j}', f'agent{unique_id}_belief_{i}', weight=weight_int_pers)  
                    G.add_edge(f'agent{unique_id}_belief_{i}', f'agent{unique_id}_belief_{j}', weight=weight_int_pers) 

        for c in all_connections:
            G.add_edge(focal_node, f'agent{unique_id}_soc_belief_{c}', weight = weight_int_soc)
            G.add_edge(f'agent{unique_id}_soc_belief_{c}', focal_node, weight = weight_int_soc)

        return G
    

    

    def step(self):
        self.select_conversion_function(self.conversion)
        self.create_full_connected_network(self.pers_belief_network, self.connected_agents, self.remaining_agents)
        self.update_state()
        
        
        
    def select_conversion_function(self, conversion):
        
        switcher = {
            "V": self.v_conversion,
            "hill": self.hill_conversion,
            "decay": self.exp_conversion
        }
        
        
        self.conversion_function = switcher.get(conversion, lambda: "Invalid conversion")
        
        
    def get_value_dict(self):
        
        G = self.pers_belief_network
        beta_dict = self.beta_intSoc_dynamic
        
        upper_limit = self.beta_intSoc*3
        
    
        focal_node = list(G.nodes(data=True))[0][1]['value']

        soc_beliefs = {int(index.split('_')[3]): belief['value'] for index, belief in G.nodes(data=True) if 'soc_belief' in index}

        for key, value in soc_beliefs.items():

            diff = abs(focal_node - value)
            #CHANGE HERE for different conversion functions
            norm = self.conversion_function(diff)

            
            current_value = beta_dict[key][-1]
            increment = (norm - 1)  * current_value * (1 - current_value / upper_limit)
            current_value += increment
            current_value = min(current_value, upper_limit)
            
            beta_dict[key].append(current_value)

        self.beta_intSoc_dynamic = beta_dict
        

    def dynamic_betas(self, adj_matrix):
        
        G = self.pers_belief_network
        beta_dict = self.beta_intSoc_dynamic
        adj = adj_matrix
    
        soc_belief_coords = {int(belief.split('_')[3]): index for index, belief in enumerate(G.nodes()) if 'soc_belief' in belief}
        

        for key, value in soc_belief_coords.items():
            #print(key, value, beta_dict)
            adj[value][0] = beta_dict[key][-1]
            adj[0][value] = beta_dict[key][-1]


        return adj
    
    
    
    def exp_conversion(self, x):
        
        a = 1.5            # starting point
        b = 1.0            # decay parameter, the higher, the stronger the decay
        y = self.exponential_decay(x, a, b)
        
        return y
        
        
    def v_conversion(self, x):
        
        a = 1.5     
        b = 1.0
        
        if x <= 1:
            y = self.exponential_decay(x, a, b)
        else:
            mirrored_val = 2 - x
            y = self.exponential_decay(mirrored_val, a, b)

        return y
        
        
        
    def hill_conversion(self, x):
        
        a = 0.6     
        b = 1
        
        if x <= 1:
            y = self.exponential_rise(x, a, b)
        else:
            mirrored_val = 2 - x
            y = self.exponential_rise(mirrored_val, a, b)

        return y
    
    
    
    def exponential_decay(self, x, a, b):

        return a * np.exp(-b * x)
    
    def exponential_rise(self, x, a, b):

        return a * np.exp(b * x)
    
    
    
    
    def create_full_connected_network(self, focal_network, connected_agents, remaining_agents):
        
        all_connections = connected_agents + remaining_agents
        
        soc_beliefs_networks = [(connected_agent.unique_id, connected_agent.pers_belief_network.subgraph([f'agent{connected_agent.unique_id}_focal_belief', f'agent{connected_agent.unique_id}_soc_belief_{self.unique_id}'])) for connected_agent in self.model.schedule.agents if connected_agent.unique_id in all_connections]
        #subgraph([f'agent{connected_agent.unique_id}_focal_belief', f'agent{connected_agent.unique_id}_soc_belief_{self.unique_id}'])
        
        tmp = []
        
        for i, n in soc_beliefs_networks:
            for e in n.nodes(data = True):
                #if 'focal' in e:
                tmp.append((i,e))
                               
        
        
        tmp_net = focal_network
        
        

        for id, net in soc_beliefs_networks:
            
            tmp_net = nx.compose(tmp_net, net)
            tmp_net.add_edge(f'agent{id}_focal_belief', f'agent{self.unique_id}_soc_belief_{id}', weight = self.weight_ext)
            tmp_net.add_edge(f'agent{self.unique_id}_focal_belief', f'agent{id}_soc_belief_{self.unique_id}', weight = self.weight_ext)
            #self.testnet = tmp_net

        
        self.fullConnNet = tmp_net





    def update_state(self):

        fullNet = self.fullConnNet
        pers_net_nodes = self.pers_belief_network.nodes()

        fullNet_adj = nx.adjacency_matrix(fullNet).toarray()
        fullNetSim = fullNet_adj.copy()
     
        fullNetSim = self.dynamic_betas(fullNetSim)
        
        fullNetSim[fullNet_adj == 0.4] = fullNet_adj[fullNet_adj == 0.4] * self.beta_intPers
        #fullNetSim[fullNet_adj == 1.0] = fullNet_adj[fullNet_adj == 1.0] * self.beta_intSoc
        fullNetSim[fullNet_adj == 1.4] = fullNet_adj[fullNet_adj == 1.4] * self.beta_ext
        
        
        #self.testnet = (fullNetSim, pers_net_nodes)
        self.get_value_dict()

        
        for index, node in enumerate(pers_net_nodes):
            
            #infos = []
            
            #Option 1: significantly more efficient
            
            connections = fullNetSim[:,index]#np.delete(fullNetSim[index], index)
            currentState = np.full((7, 1), fullNet.nodes[node]['value']) * np.ones_like(connections)

            flippedState = (np.linspace(-1, 1, 7).reshape(-1, 1)) * np.ones_like(connections)

            connectedStates = [data['value'] for _, data in fullNet.nodes(data=True)]
            connectedStates = np.array(connectedStates)#np.delete(np.array(connectedStates), index)
                
            probUnstPerValue = 1/(1 + np.exp((-np.sum(connections * flippedState * connectedStates, axis=1) + np.sum(connections * currentState * connectedStates, axis=1))))
            
            
            """
            #Option 2: original method
            probUnstPerValue = []
            for k in range(7):
                connections = fullNetSim[:,index]#np.delete(fullNetSim[index], index)
                currentState = fullNet.nodes[node]['value']
                flippedState = np.linspace(-1, 1, 7)[k]
                connectedStates = [data['value'] for _, data in fullNet.nodes(data=True)]
                connectedStates = np.array(connectedStates)#np.delete(np.array(connectedStates), index)

                probUnstPerValue = 1/(1 + np.exp((-np.sum(connections * flippedState * connectedStates, axis=1) + np.sum(connections * currentState * connectedStates, axis=1))))
                probUnstPerValue.append(prob)
                
            """

            probPerValue = [k / sum(probUnstPerValue) for k in probUnstPerValue]
            
            
            
            a = np.random.choice(np.linspace(-1, 1, 7), 1,  p=probPerValue)[0]
            fullNet.nodes[node]['value'] = a
            
            
        updated_pers_network = fullNet.subgraph(pers_net_nodes).copy()
        self.pers_belief_network = updated_pers_network
        
        

        
        
    def get_focal_node_states(self):
        
        focal_node = self.pers_belief_network.nodes()[f'agent{self.unique_id}_focal_belief']['value']

        return focal_node
    
    
    def get_mean_node_states(self):

        pattern = r'agent(\d+)_belief_(\d+)'
        
        nodes = [data['value'] for i, data in self.pers_belief_network.nodes(data=True) if re.match(pattern, i)]

        mean_node_states = (sum(nodes)/len(nodes))

        return mean_node_states
    
    def get_node_states(self):
        
        #nodes = [(i, data['value']) for i, data in self.pers_belief_network.nodes(data=True)]
        nodes = self.pers_belief_network
        
        return nodes
    
    def get_extNet(self):
        
        return self.extNet
    
    
    def get_dynamic_beta(self):

        soc_beliefs = {int(index.split('_')[3]): belief['value']  for index, belief in self.pers_belief_network.nodes(data=True) if 'soc_belief' in index}
        
        latest_beta = {key : (soc_beliefs[key], value[-1]) for key, value in self.beta_intSoc_dynamic.items()}
            
        return latest_beta
            
        
    
    def get_test(self):
        
        return self.fullConnNet
    
    def get_conn_agents(self):
        
        return self.connected_agents
   
```

```python
class NetworkModel(Model):   
    
    def __init__(self, n_nodes_int_pers, n_nodes_soc, beta_variations, ext_net, rep, conversion):

        super().__init__()

        self.schedule = RandomActivation(self)
        self.n_nodes_soc = n_nodes_soc
        self.n_nodes_int_pers = n_nodes_int_pers
        self.beta_variations = beta_variations
        self.ext_net = ext_net
        self.rep = rep    
        

        def create_belief_network(self, unique_id, n_nodes, connected_agents, remaining_agents, weight_int_pers, weight_int_soc, weight_weak_ties):


            seed = unique_id#os.getpid() + random.randint(1, 100)
            rng = default_rng(seed)
            
            #remaining_agents = [i for i in all_agents if (i not in connected_agents) | (i != unique_id)]
            
            all_agents = connected_agents + remaining_agents

            G = nx.DiGraph()

            for i in range(n_nodes):
                if i == 0: 
                    G.add_node(f'agent{unique_id}_focal_belief', value=rng.choice(np.linspace(-1, 1, 7))) 
                else:
                    G.add_node(f'agent{unique_id}_belief_{i}', value=rng.choice(np.linspace(-1, 1, 7)))
            
            for a in all_agents:
                G.add_node(f'agent{unique_id}_soc_belief_{a}', value=rng.choice(np.linspace(-1, 1, 7)))
            """
            for a in connected_agents:
                G.add_node(f'agent{unique_id}_soc_belief_{a}', value=rng.choice(np.linspace(-1, 1, 7)))           
            """


            focal_node = f'agent{unique_id}_focal_belief'

            for i in range(1, n_nodes):
                G.add_edge(focal_node, f'agent{unique_id}_belief_{i}', weight=weight_int_pers)  
                G.add_edge(f'agent{unique_id}_belief_{i}', focal_node, weight=weight_int_pers)
                for j in range(1, n_nodes):
                    if i != j:
                        G.add_edge(f'agent{unique_id}_belief_{j}', f'agent{unique_id}_belief_{i}', weight=weight_int_pers)  
                        G.add_edge(f'agent{unique_id}_belief_{i}', f'agent{unique_id}_belief_{j}', weight=weight_int_pers) 
             
            
            for c in all_agents:
                G.add_edge(focal_node, f'agent{unique_id}_soc_belief_{c}', weight = weight_int_soc)
                G.add_edge(f'agent{unique_id}_soc_belief_{c}', focal_node, weight = weight_int_soc)
                        
            
            """
            for c in connected_agents:
                G.add_edge(focal_node, f'agent{unique_id}_soc_belief_{c}', weight = weight_int_soc)
                G.add_edge(f'agent{unique_id}_soc_belief_{c}', focal_node, weight = weight_int_soc)
            """
            

            return G



        beta_soc = {}
        
        
        # Initialize agents and assign to external network nodes
        for i, node in enumerate(self.ext_net.nodes()):
            connected_agents = list(self.ext_net.neighbors(i))
            remaining_agents = [r for r in self.ext_net.nodes() if (r not in connected_agents) & (r != i)]
            beta = self.beta_variations
            #for c in connected_agents:
            beta_soc = {c:[self.beta_variations['intSoc']] for c in connected_agents}
            beta_soc.update({c:[self.beta_variations['intSoc']*0.1] for c in remaining_agents})
            network = 0 #create_belief_network(self, i, self.n_nodes_int_pers, connected_agents, remaining_agents, weight_int_pers=0.4, weight_int_soc=1.0, weight_weak_ties = 0.1)
            agent = SocialAgent(i, self, self.n_nodes_int_pers, connected_agents, remaining_agents, self.beta_variations, beta_soc, self.rep, self.ext_net, network, conversion, weight_ext=1.4)
            self.schedule.add(agent)
        
        self.datacollector = DataCollector(
            agent_reporters={"betaIntPers" : "beta_intPers"
                             , "betaIntSoc" : "beta_intSoc"
                             , "betaExt" : "beta_ext"
                             , "rep" : "rep"
                             , "focalNodeStates" : SocialAgent.get_focal_node_states
                             , "meanNodeStates" : SocialAgent.get_mean_node_states
                             #, "NodeStates" : SocialAgent.get_node_states
                             , "extNet" : SocialAgent.get_extNet
                             #, "full_net" : SocialAgent.get_test
                             , "connected_agents" : SocialAgent.get_conn_agents
                             #, "testnet" : "testnet"
                             #, "report" : "report"
                             , "dynamic_betas" : SocialAgent.get_dynamic_beta
                            })
            #model_reporters={"focalNodeStates" : self.get_focal_node_states,
            #                 "meanNodeStates" : self.get_mean_node_states}
                             
                

    def step(self):
        #if (self.schedule.steps == 1) | (self.schedule.steps == 3) | (self.schedule.steps == 5) | ((self.schedule.steps % 10) == 9):
        self.datacollector.collect(self)
        self.schedule.step()

  
```

```python
def run_simulation(params):

    model = NetworkModel(params['n_nodes_int_pers'], params['n_nodes_soc'], params['beta_variations'], params['ext_net'], params['n_reps'], params['conversion'])
    #while model.schedule.steps < params['n_steps']:
    for _ in range(params['n_steps']):
        model.step()

    # Extract and return simulation data
    agent_data = model.datacollector.get_agent_vars_dataframe()
    #model_data = model.datacollector.get_model_vars_dataframe()

    #data = pd.concat([agent_data, model_data], axis = 1)
    return agent_data





def define_param_list(n_reps, n_nodes_int_pers, n_nodes_soc, n_steps, beta_values, conversion):
    
    sizes = [n_nodes_soc // 2, n_nodes_soc // 2]
    p_matrix = [[0.4, 0.01], [0.01, 0.4]]
    ext_net = nx.stochastic_block_model(sizes, p_matrix)
    
    #sizes = [n_nodes_soc // 4, n_nodes_soc // 4, n_nodes_soc // 4, n_nodes_soc // 4]
    #p_matrix = [[0.4, 0.01, 0.01, 0.01], [0.01, 0.4, 0.01, 0.01], [0.01, 0.01, 0.4, 0.01], [.01, .01, .01, 0.4]]
    #ext_net = nx.stochastic_block_model(sizes, p_matrix, directed = False)
    
    #ext_net_adj = nx.adjacency_matrix(ext_net).toarray()
    #np.save(f"data/extNet{n_nodes_soc}_{n_steps}.npy", ext_net_adj)

        
    #with reps:
    params = [{'ext_net': ext_net,'n_nodes_int_pers' : n_nodes_int_pers, 'n_nodes_soc' : n_nodes_soc, 'n_steps' : n_steps, 'n_reps' : i, 'conversion': conversion, 'beta_variations' : {'intPers': p, 'intSoc': s, 'ext': e}} for i in range(n_reps) for p in beta_values for s in beta_values for e in beta_values]

    #params = [{'ext_net': ext_net,'n_nodes_int_pers' : n_nodes_int_pers, 'n_nodes_soc' : n_nodes_soc, 'n_steps' : n_steps, 'n_reps' : i, 'beta_variations' : {'intPers': p, 'ext': e}} for i in range(n_reps) for p in beta_values for e in beta_values for ext_net in extNets]

    return params


def pool_simulation(n_cores, params):
    sim_results= []

    pool = Pool(n_cores)

    for result in tqdm(
        pool.imap_unordered(func=run_simulation, iterable=params),
        total=len(params)
        ):
            #result['durchgang'] = params['n_reps']
            sim_results.append(result)

    pool.close()

    return pd.concat(sim_results)
```

```python
def scaled_logistic(x, L, k, x0, ymin, ymax):
    # Logistic function
    y = L / (1 + np.exp(-k * (x - x0)))
    # Scale y to be in the range [ymin, ymax]
    scaled_y = (y - np.min(y)) / (np.max(y) - np.min(y)) * (ymax - ymin) + ymin
    # Invert y around the midpoint of ymin and ymax to start and end high
    return ymax + ymin - scaled_y

# Create a sequence of x values from 0 to 2
x = np.linspace(0, 2, 20)

# Define parameters for the logistic function
L = 1        # Max value of the unscaled logistic function
k = 3      # Adjusted steepness of the curve, reduced for a wider transition
x0 = 0.1       # Midpoint of the sigmoid, where the lowest point is expected
ymin = 0.1   # Minimum y value in the desired output
ymax = 1.5     # Maximum y value in the desired output

# Calculate y values using the scaled logistic function
y = scaled_logistic(x, L, k, x0, ymin, ymax)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o')  # Line plot with points
plt.title("Adjusted Scaled and Inverted Logistic Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.savefig(f'figures/conversion/_decay_.svg', format='svg')

plt.show()
```

```python
def suppress_specific_warnings():
    warnings.filterwarnings("ignore", message=".*NoneType.*split.*")
```

```python
def get_color(value):
    # Define RGB for sea-blue and blood-red
    sea_blue = (39, 100, 123)  # RGB for #27647B
    blood_red = (202, 53, 66)  # RGB for #CA3542
    
    # Interpolate between sea-blue and blood-red based on value
    r = int(sea_blue[0] + (blood_red[0] - sea_blue[0]) * (value + 1) / 2)
    g = int(sea_blue[1] + (blood_red[1] - sea_blue[1]) * (value + 1) / 2)
    b = int(sea_blue[2] + (blood_red[2] - sea_blue[2]) * (value + 1) / 2)
    
    return f"#{r:02x}{g:02x}{b:02x}"
```

```python
def cluster_network_by_weight(network):
    
    os.environ["OMP_NUM_THREADS"] = "1"
    os.environ["MKL_NUM_THREADS"] = "1"
    
    @contextlib.contextmanager
    def suppress_stdout_stderr():
        with open(os.devnull, 'w') as devnull:
            old_stdout = sys.stdout
            old_stderr = sys.stderr
            sys.stdout = devnull
            sys.stderr = devnull
            try:
                yield
            finally:
                sys.stdout = old_stdout
                sys.stderr = old_stderr
                
                
    with suppress_stdout_stderr():

        adj_matrix = nx.to_numpy_array(network, weight='weight')
        #adj_matrix[0]


        n_clusters = 2
        spectral_clustering = SpectralClustering(n_clusters=n_clusters, affinity='precomputed', assign_labels='kmeans')
        clusters = spectral_clustering.fit_predict(adj_matrix)




        new_clusters = []
        for node, cluster in zip(network.nodes(), clusters):
            new_clusters.append((node, cluster))

        nc = pd.DataFrame(new_clusters, columns = ['agent', 'cluster'])
    
    return nc
```

```python
def get_largest_cluster_size(df):
    
    cluster_counts = df['cluster'].value_counts()

    max_agents_in_cluster = cluster_counts.max()
    
    return max_agents_in_cluster
```

```python
def process_step(df, step):
    cluster_counts = df['cluster'].value_counts().to_dict()
    cluster_0_count = cluster_counts.get(0, 0)
    cluster_1_count = cluster_counts.get(1, 0)
    return pd.DataFrame({'step': [step], 'cluster_0_count': [cluster_0_count], 'cluster_1_count': [cluster_1_count]})
```

```python
def create_results(cluster_dict):

    res_dict = {}

    for config, _dict in cluster_dict.items():

        analysis_columns = ['step', 'cluster_0_count', 'cluster_1_count']
        analysis = pd.DataFrame(columns=analysis_columns)

        for step, network in tqdm(_dict.items()):

            df = consistent_clustering(cluster_network_by_weight(network))
            result = process_step(df, step)
            analysis = pd.concat([analysis, result], ignore_index=True)

        res_dict[config] = analysis
        
    return res_dict
    
```

```python
def create_dict(data, steps):
    
    #values = [0.5, 2]
    #combinations = list(itertools.product(values, values, values))
    combinations = data.formatted.unique()

    conn_dict = {}
    for formatted in combinations:
        res_dict = {}
        for i in tqdm(range(steps)):
            res_dict[i] = create_cluster_network(data, i, formatted)

        conn_dict[f"{formatted}"] = res_dict
        
    return conn_dict
```

```python
def create_cluster_network(data, step, formatted):
    
    suppress_specific_warnings()
    
    #p = pers   # beta pers variation, currently 2 or .5
    #i = soc  # beta soc variation, currently 2 or .5
    #e = ext   # beta ext variation, currently 2 or .5

    df = data[(data.betaIntSoc == formatted)].copy()
    df_step = df[df.Step == step]

    edges_in_graph = []
    for index, row in df_step.iterrows():
        d = row['dynamic_betas']
        for key, value in d.items():
            edges_in_graph.append((row.AgentID, key, value[1]))

    G = nx.Graph()

    grouped = df_step.groupby('AgentID')['focalNodeStates'].mean()
    for agent, focalNodeState in grouped.items():
        node_color = get_color(focalNodeState)

        G.add_node(int(agent), value=focalNodeState)

    #calculate mean weights for mutliple repitition runs
    edge_weights = defaultdict(lambda: {'sum': 0, 'count': 0})

    for u, v, weight in edges_in_graph:
        edge_weights[(u, v)]['sum'] += weight
        edge_weights[(u, v)]['count'] += 1

    mean_edge_weights = {edge: data['sum'] / data['count'] for edge, data in edge_weights.items()}



    #define edge intensity of mean weighted edges
    for edge, weight in mean_edge_weights.items():
        G.add_edge(edge[0], edge[1], weight=weight)
        
    return G
```

```python
def create_connection_columns(df):
    
    for row in df.itertuples(index=True):
        friends = []
        new_enemies = []
        old_friends = []
        new_friends = []
        different_cluster = []

        for key, value in row.dynamic_betas.items():
            if row.AgentID < 50:
                if value[1] >= row.betaIntSoc*0.2:
                    friends.append(key)
                    if key in row.connected_agents:
                        old_friends.append(key)
                    elif key >= 50:
                        different_cluster.append(key)
                    else:
                        new_friends.append(key)
                elif value[1] < row.betaIntSoc*0.1:
                    if key in row.connected_agents:
                        new_enemies.append(key)
                        
            if row.AgentID >= 50:
                if value[1] >= row.betaIntSoc*0.2:
                    friends.append(key)
                    if key in row.connected_agents:
                        old_friends.append(key)
                    elif key < 50:
                        different_cluster.append(key)
                    else:
                        new_friends.append(key)
                elif value[1] < row.betaIntSoc*0.1:
                    if key in row.connected_agents:
                        new_enemies.append(key)


        df.at[row.Index, 'connections'] = friends
        df.at[row.Index, 'disconnected'] = new_enemies
        df.at[row.Index, 'established_connections'] = old_friends
        df.at[row.Index, 'new_connections'] = new_friends
        df.at[row.Index, 'inter_cluster_connection'] = different_cluster
        
    return df
```

```python
def plot_network_wo_outliers(data, steps, pers, soc, ext):
    
    p = pers   # beta pers variation, currently 2 or .5
    i = soc  # beta soc variation, currently 2 or .5
    e = ext   # beta ext variation, currently 2 or .5

    fig, axs = plt.subplots(1, len(steps), figsize=(24, 12))
    gs = GridSpec(2, len(steps), height_ratios=[1, 3])

    for idx, step in enumerate(steps):
        try:
            df = data[(data.betaIntSoc == i) & (data.betaIntPers == p) & (data.betaExt == e) & (data.rep == 0)].copy()
            df_step = df[df.Step == step]

            edges_in_graph = []
            for index, row in df_step.iterrows():
                d = row['dynamic_betas']
                for key, value in d.items():
                    edges_in_graph.append((row.AgentID, key, value[1]))

            G = nx.Graph()

            for index, row in df_step.iterrows():
                node_color = get_color(row['focalNodeStates'])

                if row['AgentID'] < 50:
                    node_shape = 's'  # square
                    node_size = 200
                    origin_cluster = 0

                else:
                    node_shape = '*'  # star
                    node_size = 400
                    origin_cluster = 1

                G.add_node(int(row['AgentID']), color=node_color, shape=node_shape, size=node_size)

            for start, end, weight in edges_in_graph:
                if weight >= i*0.2:
                    intensity = 150 - int(weight)
                    color = f"#{intensity:02x}{intensity:02x}{intensity:02x}"
                    G.add_edge(start, end, weight=weight, color=color)



            pos = nx.kamada_kawai_layout(G, weight='weight')

            ###################### disregard outliers ##############################
            
            edge_lengths = {}
            for u, v in G.edges():
                dist = np.linalg.norm(np.array(pos[u]) - np.array(pos[v]))
                G[u][v]['length'] = dist
                if u not in edge_lengths:
                    edge_lengths[u] = []
                if v not in edge_lengths:
                    edge_lengths[v] = []
                edge_lengths[u].append(dist)
                edge_lengths[v].append(dist)

            mean_distances = {node: np.mean(lengths) for node, lengths in edge_lengths.items()}


            distance_threshold = np.percentile(list(mean_distances.values()), 90)

            filtered_nodes = {node for node, mean_dist in mean_distances.items() if mean_dist < distance_threshold}
            filtered_edges = [(u, v, G[u][v]) for u, v in G.edges if u in filtered_nodes and v in filtered_nodes]

            G_filtered = nx.Graph()
            G_filtered.add_nodes_from((node, G.nodes[node]) for node in filtered_nodes)
            G_filtered.add_edges_from(filtered_edges)

            pos_filtered = nx.kamada_kawai_layout(G_filtered, weight='weight') 
            
            edge_lengths = {}
            for u, v in G_filtered.edges():
                dist = np.linalg.norm(np.array(pos[u]) - np.array(pos[v]))
                G[u][v]['length'] = dist
                if u not in edge_lengths:
                    edge_lengths[u] = []
                if v not in edge_lengths:
                    edge_lengths[v] = []
                edge_lengths[u].append(dist)
                edge_lengths[v].append(dist)

            mean_distances = {node: np.mean(lengths) for node, lengths in edge_lengths.items()}
            


            G_real_conn = nx.Graph()
            real_conn_edges = [(u, v, w['weight']) for u, v, w in G_filtered.edges(data=True) if w['weight'] > 0.1]


            G_real_conn.add_weighted_edges_from(real_conn_edges)
            degrees_real_conn = dict(G_real_conn.degree(weight=None))

            try:
                mean_degree_real_conn = sum(degrees_real_conn.values()) / len(degrees_real_conn)
            except:
                mean_degree_real_conn = 0

            weighted_degrees = dict(G_filtered.degree(weight=None))

            try:
                weighted_mean_degree = sum(weighted_degrees.values()) / len(weighted_degrees)

            except:
                weighted_mean_degree = 0




            ax_net = fig.add_subplot(gs[1, idx])
            ax_net.set_title(f'Step {step}')

            node_shapes = {'s', 'o', '^', '*'}
            for shape in node_shapes:
                nodes_of_shape = [node for node in G_filtered.nodes if G_filtered.nodes[node]['shape'] == shape]
                sizes_of_shape = [G_filtered.nodes[node]['size'] for node in nodes_of_shape]
                colors_of_shape = [G_filtered.nodes[node]['color'] for node in nodes_of_shape]

                nx.draw_networkx_nodes(
                    G_filtered,
                    pos_filtered,
                    nodelist=nodes_of_shape,
                    node_shape=shape,
                    node_color=colors_of_shape,
                    node_size=sizes_of_shape,
                    ax=ax_net
                )

            edges_colors = [G_filtered.edges[edge]['color'] for edge in G_filtered.edges]
            edges_widths = [G_filtered.edges[edge]['weight'] for edge in G_filtered.edges]
            nx.draw_networkx_edges(G_filtered, pos_filtered, edge_color=edges_colors, width=edges_widths, ax=ax_net)

            #ax_net.legend(handles=[
            #    mpatches.Patch(color=get_color(-1), label='Opinion: -1'),
            #    mpatches.Patch(color=get_color(-0.66), label='Opinion: -0.66'),
            #    mpatches.Patch(color=get_color(-0.33), label='Opinion: -0.33'),
            #    mpatches.Patch(color=get_color(0), label='Opinion: 0'),
            #    mpatches.Patch(color=get_color(0.33), label='Opinion: 0.33'),
            #    mpatches.Patch(color=get_color(0.66), label='Opinion: 0.66'),
            #    mpatches.Patch(color=get_color(1), label='Opinion: 1')
            #], title='Opinion State', loc='lower left')

            ax_net.annotate(f'Mean Degree: {weighted_mean_degree:.2f}', xy=(0.05, 0.95), xycoords='axes fraction',
                        fontsize=12, color='blue', bbox=dict(boxstyle="round,pad=0.3", edgecolor='blue', facecolor='white'))
            ax_net.annotate(f'Filtered Mean Degree: {mean_degree_real_conn:.2f}', xy=(0.95, 0.95), xycoords='axes fraction',
                        fontsize=12, color='red', ha='right', bbox=dict(boxstyle="round,pad=0.3", edgecolor='red', facecolor='white'))

            # Plot the histogram of mean distances
            ax_hist = fig.add_subplot(gs[0, idx])
            ax_hist.hist(mean_distances.values(), bins=20, color='gray', edgecolor='black')
            ax_hist.set_title(f'Step {step+1} - Distances Histogram')
            ax_hist.set_ylim(0, 40)
            ax_hist.set_xlim(0.2, 0.9)
            ax_hist.set_xlabel('Mean Distance')
            ax_hist.set_ylabel('Frequency')
        
        except:
            continue


    fig.suptitle(f'Weighted network with starting beta social = {i} | beta personal = {p} | beta external = {e}')
    plt.tight_layout()
    plt.savefig(f'figures/_{conversion}_/network_wo_outliers_{p}_{i}_{e}_.svg', format='svg')
    plt.show()
    
    #return sorted_mean_distances, G
```

```python
def create_results_df(df):

    n = len(df)
    quarter_size = n // 2

    df['initial_cluster'] = (df.index // quarter_size)

    results = {
        'initial_cluster': [],
        'initial_count': [],
        'current_count': [],
        'joined': [],
        'left': [],
        'ratio_of_original_members': []
    }

    for cluster in range(2):
        initial_agents = df[df['initial_cluster'] == cluster]

        predominant_cluster = Counter(initial_agents['new_cluster']).most_common(1)[0][0]


        initial_count = len(initial_agents)
        current_count = len(df[df['new_cluster'] == predominant_cluster])
        joined = current_count - len(initial_agents[initial_agents['new_cluster'] == predominant_cluster])
        left = initial_count - len(initial_agents[initial_agents['new_cluster'] == predominant_cluster])
        ratio = len(initial_agents[initial_agents['new_cluster'] == predominant_cluster]) / initial_count


        results['initial_cluster'].append(cluster)
        results['initial_count'].append(initial_count)
        results['current_count'].append(current_count)
        results['joined'].append(joined)
        results['left'].append(left)
        results['ratio_of_original_members'].append(ratio)


    results_df = pd.DataFrame(results)
    
    return results_df
```

```python
def create_belief_network(unique_id, n_nodes, connected_agents, weight_int_pers, weight_int_soc):

    seed = os.getpid() + unique_id
    #np.random.seed(seed)
    rng = default_rng(seed)

    G = nx.DiGraph()

    for i in range(n_nodes):
        if i == 0: 
            G.add_node(f'agent{unique_id}_focal_belief', value=rng.choice(np.linspace(-1, 1, 7))) 
        else:
            G.add_node(f'agent{unique_id}_belief_{i}', value=rng.choice(np.linspace(-1, 1, 7)))

    for c in connected_agents:
        G.add_node(f'agent{unique_id}_soc_belief_{c}', value=rng.choice(np.linspace(-1, 1, 7)))

        #print(f'agent{unique_id}_soc_belief_{c}')


    focal_node = f'agent{unique_id}_focal_belief'

    for i in range(1, n_nodes):
        G.add_edge(focal_node, f'agent{unique_id}_belief_{i}', weight=weight_int_pers)  
        G.add_edge(f'agent{unique_id}_belief_{i}', focal_node, weight=weight_int_pers)
        for j in range(1, n_nodes):
            if i != j:
                G.add_edge(f'agent{unique_id}_belief_{j}', f'agent{unique_id}_belief_{i}', weight=weight_int_pers)  
                G.add_edge(f'agent{unique_id}_belief_{i}', f'agent{unique_id}_belief_{j}', weight=weight_int_pers) 

    for c in connected_agents:
        G.add_edge(focal_node, f'agent{unique_id}_soc_belief_{c}', weight = weight_int_soc)
        G.add_edge(f'agent{unique_id}_soc_belief_{c}', focal_node, weight = weight_int_soc)


    return G
```

```python
def create_cluster_network(data, step, pers, soc, ext):
    
    suppress_specific_warnings()
    
    p = pers   # beta pers variation, currently 2 or .5
    i = soc  # beta soc variation, currently 2 or .5
    e = ext   # beta ext variation, currently 2 or .5

    df = data[(data.betaIntSoc == i) & (data.betaIntPers == p) & (data.betaExt == e)].copy()
    df_step = df[df.Step == step]

    edges_in_graph = []
    for index, row in df_step.iterrows():
        d = row['dynamic_betas']
        for key, value in d.items():
            edges_in_graph.append((row.AgentID, key, value[1]))

    G = nx.Graph()

    grouped = df_step.groupby('AgentID')['focalNodeStates'].mean()
    for agent, focalNodeState in grouped.items():
        node_color = get_color(focalNodeState)

        if agent < 50:
            origin_cluster = 0

        else:
            origin_cluster = 1

        G.add_node(int(agent), value=focalNodeState, cluster = origin_cluster)

    #calculate mean weights for mutliple repitition runs
    edge_weights = defaultdict(lambda: {'sum': 0, 'count': 0})

    for u, v, weight in edges_in_graph:
        edge_weights[(u, v)]['sum'] += weight
        edge_weights[(u, v)]['count'] += 1

    mean_edge_weights = {edge: data['sum'] / data['count'] for edge, data in edge_weights.items()}



    #define edge intensity of mean weighted edges
    for edge, weight in mean_edge_weights.items():
        G.add_edge(edge[0], edge[1], weight=weight)
        
    return G
```

```python
def scaled_logistic(x, L, k, x0, ymin, ymax):
    # Logistic function
    y = L / (1 + np.exp(-k * (x - x0)))
    # Scale y to be in the range [ymin, ymax]
    scaled_y = (y - np.min(y)) / (np.max(y) - np.min(y)) * (ymax - ymin) + ymin
    # Invert y around the midpoint of ymin and ymax to start and end high
    return ymax + ymin - scaled_y

# Create a sequence of x values from 0 to 2
x = np.linspace(0, 2, 20)

# Define parameters for the logistic function
L = 1        # Max value of the unscaled logistic function
k = 3      # Adjusted steepness of the curve, reduced for a wider transition
x0 = 0.5       # Midpoint of the sigmoid, where the lowest point is expected
ymin = 0.1   # Minimum y value in the desired output
ymax = 1.5     # Maximum y value in the desired output

# Calculate y values using the scaled logistic function
y = scaled_logistic(x, L, k, x0, ymin, ymax)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o')  # Line plot with points
plt.title("Adjusted Scaled and Inverted Logistic Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
def create_belief_network(unique_id, n_nodes, connected_agents, weight_int_pers = 0.4, weight_int_soc = 1.0):

    seed = os.getpid() + unique_id
    #np.random.seed(seed)
    rng = default_rng(seed)

    G = nx.DiGraph()

    for i in range(n_nodes):
        G.add_node(f'agent{unique_id}_belief_{i}', value=rng.choice(np.linspace(-1, 1, 7)))

    for c in connected_agents:
        G.add_node(f'agent{unique_id}_soc_belief_{c}', value=rng.choice(np.linspace(-1, 1, 7)))


    focal_node = f'agent{unique_id}_belief_0'

    for i in range(1, n_nodes):
        G.add_edge(focal_node, f'agent{unique_id}_belief_{i}', weight=weight_int_pers)  
        G.add_edge(f'agent{unique_id}_belief_{i}', focal_node, weight=weight_int_pers)
        for j in range(1, n_nodes):
            if i != j:
                G.add_edge(f'agent{unique_id}_belief_{j}', f'agent{unique_id}_belief_{i}', weight=weight_int_pers)  
                G.add_edge(f'agent{unique_id}_belief_{i}', f'agent{unique_id}_belief_{j}', weight=weight_int_pers) 

    for c in connected_agents:
        G.add_edge(focal_node, f'agent{unique_id}_soc_belief_{c}', weight = weight_int_soc)
        G.add_edge(f'agent{unique_id}_soc_belief_{c}', focal_node, weight = weight_int_soc)


    return G
```

```python
def consistent_clustering(df):
    cluster_means = df.groupby('cluster')['agent'].mean()

    cluster_mapping = {0: cluster_means.idxmin(), 1: cluster_means.idxmax()}
    df['cluster'] = df['cluster'].map(cluster_mapping)
    
    # get the number of agents in the biggest cluster
    cluster_counts = df['cluster'].value_counts()

    max_agents_in_cluster = cluster_counts.max()
    
    return max_agents_in_cluster
```
